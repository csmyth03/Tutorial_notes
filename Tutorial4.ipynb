# Loop Functions


## apply ()
- Data frame or matrix as input 
Syntax: 
apply(X,MARGIN, FUNCTION)

```r
iris
view(iris)

apply_mean.col <- apply(iris[,1:4],2,mean)
apply_mean.col
```

## lappyly()
- used for vector, list. 
syntax : lapply(X, FUN)
X : vector, list, matrix. 
FUN : function to be applied

Will return a list.

```r
vector.a <- c("Hello, world!","This is R programming.")
lapply.min <- lapply(vector.a, toupper)
lapply.min
```

## Sapply()
Behaves same as lapply = returns vector 
sapply (X, fun)
X can be vector, matrix, dataframe or list

```r
sapply.min <- sapply(vector.a, toupper)
sapply.min
```

# tapply - used for factor
Applies function over a subset
makes grouping based on index values
index values = attribute in data strcture, which contains factor values e.g. categorical factors like different petal types

```r
tapply.class.mean <- tapply(iris$Sepal.Length, iris$Species, mean)
tapply.class.mean
```

## split() - used to split data with a factor
vector or objects and splits into groups determind by factor


```r
iris.data.split <- split(iris.data, iris.data$class)
iris.data.split
```

# Error handling function = tryCatch()

```r
tryCatch(
    {a = as.numeric(readline("Enter the first number:"))
    b = as.numeric(readline("Enter your second number:"))
    c = sum.values(a,b)
    print(paste("The sum of the two values is:", c))}, 
    error = function(e){message("Name error: \n", e)},
    warning = function(w){message("Ensure your input variables are numeric:\n ",w)},
    finally = {message("Done!")}
)
```
